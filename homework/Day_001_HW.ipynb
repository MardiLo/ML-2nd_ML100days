### 作業3

```python
def mean_squared_error(y, yp):
    """
    計算 MSE
    Args:
        - y: 實際值
        - yp: 預測值
    Return:
        - mse: MSE
    """
    
    mse = sum((y-yp)**2)/len(y)
    return mse
```
