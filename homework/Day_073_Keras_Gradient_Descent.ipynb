# 範例重點

定義函數: function y=(x+5)² 
在 x=3 開始, 快速收斂

# 執行步驟

Step 1 : Initialize x =3. Then, find the gradient of the function, dy/dx = 2*(x+5).

Step 2 : Move in the direction of the negative of the gradient. We use a learning rate. Let us assume the learning rate → 0.01

Step 3 : Let’s perform 2 iterations of gradient descent

Step 4 : We can observe that the X value is slowly decreasing and should converge to -5 (the local minima). However, how many iterations should we perform?

```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
cur_x = 3 # The algorithm starts at x=3
lr = 0.1 # Learning rate
precision = 10**-5 # This tells us when to stop the algorithm
step_size = 1 # Arbitrarily setting step size
max_iters = 10**5 # Maximum number of iterations
iters = 0 # Iteration counter
df = lambda x: 2*(x+5) # Gradient of our function 

iters_history = [iters]
x_history = [cur_x]
```

```python
while step_size > precision and iters < max_iters:
    prev_x = cur_x # Store value in prev_x for calculating step size later
    cur_x = cur_x - lr * df(cur_x) # Gradient descent
    step_size = abs(cur_x - prev_x)
    iters += 1
    print(f"iter:{iters}, current x position:{cur_x:.5f}, step size:{step_size:.5f}")

    iters_history.append(iters)
    x_history.append(cur_x)

print()
print(f"Totally iteations: {iters_history[-1]}")
print(f"The local minimum occurs at: {x_history[-1]}")
```

```python
plt.plot(iters_history, x_history)
plt.xlabel('iters'); plt.ylabel('x');
```
