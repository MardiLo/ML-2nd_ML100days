- 作業目標 : iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # for 3D projection to work

from sklearn.cluster import KMeans
from sklearn import datasets

%matplotlib inline
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target

display(X)
display(y)
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

```python
# 設定模型參數
estimators = [('k_means_6', KMeans(n_clusters=6)),
              ('k_means_3', KMeans(n_clusters=3)),
              ('k_means_bad_init', KMeans(n_clusters=8, n_init=1, init='random'))]
```

```python
# 資料建模並視覺化結果
fignum = 1
titles = ['6 clusters', '3 clusters', '8 clusters, bad initialization']
for name, est in estimators:
    fig = plt.figure(fignum, figsize=(6, 6))
    ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
    
    ## fit data
    est.fit(X)
    
    labels = est.labels_

    ax.scatter(X[:, 3], X[:, 0], X[:, 2],
               c=labels.astype(np.float), edgecolor='k')

    ax.set_title(titles[fignum - 1])
    ax.dist = 12
    fignum = fignum + 1
```
