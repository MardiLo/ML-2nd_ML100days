```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Sigmoid

$ \sigma (x) = \frac{1}{1+e^{-x}} $

$ \frac{d\sigma (x)}{d(x)} = \sigma (x)\cdot (1-\sigma(x)) $


```python
# Sigmoid function
def sigmoid(x):
    return (1 / (1 + np.exp(-x)))

# Derivative of Sigmoid function
def dsigmoid(x):
    return (sigmoid(x) * (1 - sigmoid(x)))

x = np.linspace(-10,10,1000)

plt.grid()
plt.title('Sigmoid Function')
plt.text(2.5, 0.7, r'$\sigma(x)=\frac{1}{1+e^{-x}}$', fontsize=20)
plt.plot(x, sigmoid(x), 'b');
plt.plot(x, dsigmoid(x), 'r');
```

# Softmax
$S(x_i)=\frac{e^{x_i}}{\sum_{j}e^{x_{j}}}$

```python
# Softmax function
def softmax(x):
     return (np.exp(x) / float(sum(np.exp(x))))

x = np.linspace(-10,10,1000)

plt.grid()
plt.title('Softmax Function')
plt.text(0, 0.015, r'$S(x_i)=\frac{e^{x_i}}{\sum_{j}e^{x_{j}}}$', fontsize=20)
plt.plot(x, softmax(x), 'b');
```
