# 作業目標:
    請嘗試使用不同的組合驗證 learning rate 對所需 iteration 的影響
    lr = [0.1, 0.0001]
    主要驗證 Lr 對於grandient 收斂的速度
    

```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
rows, cols = 1, 3
figure, ax = plt.subplots(rows, cols, figsize=(16,6))

for i, lr in enumerate([0.1, 0.01, 0.0001]): # Learning rate
    cur_x = 3 # The algorithm starts at x=3
    precision = 10**-5 # This tells us when to stop the algorithm
    step_size = 1 
    max_iters = 10**5 # Maximum number of iterations
    iters = 0 # Iteration counter
    df = lambda x: 2*(x+5) # Gradient of our function 

    iters_history = [iters]
    x_history = [cur_x]

    while step_size > precision and iters < max_iters:
        prev_x = cur_x # Store value in prev_x for calculating step size later
        cur_x = cur_x - lr * df(cur_x) # Gradient descent
        step_size = abs(cur_x - prev_x)
        iters += 1

        iters_history.append(iters)
        x_history.append(cur_x)
    
    print()
    print(f"Totally iteations: {iters_history[-1]}")
    print(f"The local minimum occurs at: {x_history[-1]}")
    
    plt.subplot(rows, cols, i+1)
    plt.plot(iters_history, x_history)
    plt.xlabel('iters'); plt.ylabel('x'); plt.title(f"lr={lr}")
```
